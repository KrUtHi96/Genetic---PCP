```python
import random
import numpy as np
```

```python
dna_length = 1000
dna = ''
prob_dist = []

for _ in range(dna_length):
    curr_val = random.gauss(0, 1)
    
    if curr_val < -0.5:
        dna += 'A'
    elif curr_val < 0:
        dna += 'C'
    elif curr_val < 0.5:
        dna += 'G'
    else:
        dna += 'T'
print(dna)
```

```python
n_reads = 1000
read_length = 100
reads = list()
for i in range(n_reads):
    start_index = random.randint(0, dna_length - read_length - 1)
    reads.append(dna[start_index: start_index + read_length])
        
print(reads)
```
